## Data Preprocessing

```python
# Function to clean non-breaking spaces
def clean_non_breaking_spaces(data):
    if isinstance(data, dict):
        return {k: clean_non_breaking_spaces(v) for k, v in data.items()}
    elif isinstance(data, list):
        return [clean_non_breaking_spaces(item) for item in data]
    elif isinstance(data, str):
        return data.replace('\u00a0', ' ')
    return data


# Load, clean, and save the JSON file
with open('SJSU_courses_dataset.json', 'r', encoding='utf-8') as infile:
    courses_data = json.load(infile)

cleaned_data = clean_non_breaking_spaces(courses_data)

with open('processed_SJSU_courses_dataset.json', 'w', encoding='utf-8') as outfile:
    json.dump(cleaned_data, outfile, indent=4)

print("Non-breaking spaces removed and dataset saved to 'processed_SJSU_courses_dataset.json'")
```

```python
def ensure_list_format(value):
    if isinstance(value, dict):
        return list(value.values())
    elif not isinstance(value, list):
        return [value]
    return value

# Load the dataset
with open('processed_SJSU_courses_dataset.json', 'r', encoding='utf-8') as infile:
    courses_data = json.load(infile)

# Preprocess the dataset
processed_data = []
for idx, entry in enumerate(courses_data):
    course_entry = {
        "id": str(entry.get("id", idx + 1)),
        "title": entry.get("title", "No Title"),
        "units": entry.get("units", "N/A"),
        "description": entry.get("description", "No Description"),
        "prerequisite(s)": ensure_list_format(entry.get("prerequisite(s)", "No prerequisites listed")),
        "corequisite(s)": ensure_list_format(entry.get("corequisite(s)", "No corequisites listed")),
        "pre/corequisite(s)": ensure_list_format(entry.get("pre/corequisite(s)", "No pre/corequisites listed")),
        "grading": entry.get("grading", "N/A"),
        "notes": ensure_list_format(entry.get("note(s)", "No notes available")),
        "class_structure": entry.get("class_structure", "Class structure not found")
    }
    processed_data.append(course_entry)

# Convert to Hugging Face Dataset format
dataset = Dataset.from_dict({
    "id": [entry["id"] for entry in processed_data],
    "title": [entry["title"] for entry in processed_data],
    "units": [entry["units"] for entry in processed_data],
    "description": [entry["description"] for entry in processed_data],
    "prerequisite(s)": [entry["prerequisite(s)"] for entry in processed_data],
    "corequisite(s)": [entry["corequisite(s)"] for entry in processed_data],
    "pre/corequisite(s)": [entry["pre/corequisite(s)"] for entry in processed_data],
    "grading": [entry["grading"] for entry in processed_data],
    "notes": [entry["notes"] for entry in processed_data],
    "class_structure": [entry["class_structure"] for entry in processed_data]
})

with open('complete_SJSU_courses_dataset.json', 'w', encoding='utf-8') as outfile:
    json.dump([entry for entry in processed_data], outfile, indent=4)

print("Dataset saved to 'complete_SJSU_courses_dataset.json'")
```

# Models

```python
warnings.filterwarnings('ignore')

def load_json_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        json_data = json.load(file)

    if not json_data:
        raise ValueError("JSON data is empty")

    print(f"Successfully loaded {len(json_data)} courses")
    return json_data

# Document Processing Setup
def initialize_text_splitter():
    return RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50,
        separators=["\n\n", "\n", ".", "!", "?", ",", " ", ""]
    )

def initialize_embeddings():
    return HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2",
        model_kwargs={'device': 'cpu'}
    )

# Process course data
def process_course_data(json_data):
    documents = []
    for course in json_data:
        content = [
            f"Title: {course.get('title', 'N/A')}",
            f"Units: {course.get('units', 'N/A')}",
            f"Description: {course.get('description', 'N/A')}",
            f"Grading: {course.get('grading', 'N/A')}",
            f"Class Structure: {course.get('class_structure', 'Class structure not found')}"
        ]

        if course.get('prerequisite(s)'):
            content.append("Prerequisite(s):")
            for prereq in course['prerequisite(s)']:
                content.append(f"- {prereq}")

        if course.get('corequisite(s)'):
            content.append("Corequisite(s):")
            for coreq in course['corequisite(s)']:
                content.append(f"- {coreq}")

        if course.get('pre/corequisite(s)'):
            content.append("Pre/Corequisite(s):")
            for pre_coreq in course['pre/corequisite(s)']:
                content.append(f"- {pre_coreq}")

        if course.get('notes'):
            content.append("Note(s):")
            for note in course['notes']:
                content.append(f"- {note}")

        doc = Document(
            page_content="\n".join(content),
            metadata={"title": course.get('title', 'N/A')}
        )
        documents.append(doc)

    return documents

# Vector Store Creation
def create_vector_store(documents, text_splitter, embeddings):
    # Split documents into chunks
    splits = text_splitter.split_documents(documents)

    vector_store = FAISS.from_documents(splits, embeddings)
    return vector_store

# Model Initialization
def initialize_t5_model():
    model_name = "google/flan-t5-base"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.clean_up_tokenization_spaces = True
    model = AutoModelForSeq2SeqLM.from_pretrained(
        model_name,
        device_map="auto",
        torch_dtype=torch.float32
    )

    pipe = pipeline(
        "text2text-generation",
        model=model,
        tokenizer=tokenizer,
        max_length=512,
        temperature=0.3,
        top_p=0.95,
        do_sample=True,
        repetition_penalty=1.15
    )

    return HuggingFacePipeline(pipeline=pipe)

def initialize_bart_model():
    model_name = "facebook/bart-large-cnn"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.clean_up_tokenization_spaces = True

    model = AutoModelForSeq2SeqLM.from_pretrained(model_name, torch_dtype=torch.float32)
    device = 0 if torch.cuda.is_available() else -1
    model = model.to(device)

    pipe = pipeline(
        "text2text-generation",
        model=model,
        tokenizer=tokenizer,
        max_length=512,
        temperature=0.3,
        top_p=0.95,
        do_sample=True,
        repetition_penalty=1.15,
        device=device
    )

    return HuggingFacePipeline(pipeline=pipe)

# RAG Chain Setup
def setup_rag_chain(llm, vector_store):
    template = """[INST] Answer based on this context:
    {context}

    Question: {question}
    """

    prompt = PromptTemplate(
        template=template,
        input_variables=["context", "question"]
    )

    chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=vector_store.as_retriever(search_kwargs={"k": 2}),
        chain_type_kwargs={
            "prompt": prompt
            # "verbose": True
        },
        return_source_documents=True
    )

    return chain

def create_visa_assistant(json_data, model_type):
    text_splitter = initialize_text_splitter()
    embeddings = initialize_embeddings()
    documents = process_course_data(json_data)
    vector_store = create_vector_store(documents, text_splitter, embeddings)

    model_initializers = {
        'flan-t5': initialize_t5_model,
        'bart': initialize_bart_model
    }

    llm = model_initializers[model_type.lower()]()
    qa_chain = setup_rag_chain(llm, vector_store)

    return qa_chain

def query_visa_assistant(qa_chain, question):
    result = qa_chain.invoke({"query": question})
    return {
        "answer": result.get("result", "No response generated"),
        "source_documents": result.get("source_documents", [])
    }

def collect_model_responses(json_data, questions, model_types=['flan-t5', 'bart']):
    data = []

    # Process each question
    for i, question in enumerate(questions, 1):
        print(f"\nProcessing question {i}/{len(questions)}")
        question_data = {
            'question_id': i,
            'question': question
        }

        for model_type in model_types:
            print(f"Processing with {model_type.upper()} model...")

            start_time = time.time()
            qa_chain = create_visa_assistant(json_data, model_type)
            setup_time = time.time() - start_time

            response_start_time = time.time()
            response = query_visa_assistant(qa_chain, question)
            response_time = time.time() - response_start_time

            question_data.update({
                f'{model_type}_response': response['answer'],
                f'{model_type}_response_time': round(response_time, 2),
                f'{model_type}_response_length': len(response['answer']),
                f'{model_type}_source_documents': response['source_documents']
            })

        data.append(question_data)

    df = pd.DataFrame(data)
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.width', 1000)

    return df
```

# Evaluation

```python
# def calculate_rouge_scores(prediction: str, reference: str) -> dict:
#     scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
#     scores = scorer.score(reference, prediction)
#     return {
#         'rouge1': scores['rouge1'].fmeasure,
#         'rouge2': scores['rouge2'].fmeasure,
#         'rougeL': scores['rougeL'].fmeasure
#     }

# def calculate_bleu_score(prediction: str, reference: str) -> float:
#     smoother = SmoothingFunction()
#     prediction_tokens = word_tokenize(prediction.lower())
#     reference_tokens = [word_tokenize(reference.lower())]
#     return sentence_bleu(reference_tokens, prediction_tokens, smoothing_function=smoother.method1)

def calculate_bertscore(prediction: str, reference: str):
    P, R, F1 = score([prediction], [reference], lang="en", verbose=False)
    return F1.mean().item()

def check_answer_presence(prediction: str, source_docs: list) -> float:
    nlp = spacy.load("en_core_web_sm")
    stopwords_set = set(stopwords.words("english"))

    pred_doc = nlp(prediction)
    source_text = ' '.join(source_docs) if source_docs else ''
    source_doc = nlp(source_text)

    pred_elements = set()
    pred_elements.update([ent.text.lower() for ent in pred_doc.ents])
    pred_elements.update([chunk.text.lower() for chunk in pred_doc.noun_chunks])

    if not pred_elements:
        return 1.0

    matches = 0
    for element in pred_elements:
        if element in source_text.lower() and element not in stopwords_set:
            matches += 1

    return matches / len(pred_elements)

def check_consistency(prediction: str, source_docs: list) -> float:
    nlp = spacy.load("en_core_web_sm")
    stopwords_set = set(stopwords.words("english"))

    pred_doc = nlp(prediction)
    source_doc = nlp(' '.join(source_docs)) if source_docs else nlp('')

    pred_numbers = [token.text for token in pred_doc if token.like_num and token.text not in stopwords_set]
    source_numbers = [token.text for token in source_doc if token.like_num and token.text not in stopwords_set]

    if not pred_numbers:
        return 1.0

    matches = sum(1 for num in pred_numbers if num in source_numbers)
    return matches / len(pred_numbers) if pred_numbers else 1.0

def evaluate_models(responses_df, model_types, reference_dict):
    individual_metrics_data = {
        'model': [],
        'question_id': [],
        'bertscore_f1': [],
        'answer_presence': [],
        'consistency': []
    }

    average_metrics_data = {
        'model': [],
        'bertscore_f1': [],
        'answer_presence': [],
        'consistency': []
    }

    for model_type in model_types:
        all_metrics = {
            'bertscore_f1': [],
            'answer_presence': [],
            'consistency': []
        }

        for _, row in responses_df.iterrows():
            question_id = row['question_id']
            model_response = row[f"{model_type}_response"]
            source_documents = row[f"{model_type}_source_documents"]

            reference_answer = reference_dict.get(question_id, "")

            model_response = model_response if isinstance(model_response, str) else ''
            source_documents = ' '.join([doc.page_content if hasattr(doc, 'page_content') else str(doc) for doc in source_documents]) if isinstance(source_documents, list) else source_documents

            bert_score = calculate_bertscore(model_response, reference_answer)
            answer_presence = check_answer_presence(model_response, [source_documents])
            consistency = check_consistency(model_response, [source_documents])

            individual_metrics_data['model'].append(model_type.upper())
            individual_metrics_data['question_id'].append(question_id)
            individual_metrics_data['bertscore_f1'].append(bert_score)
            individual_metrics_data['answer_presence'].append(answer_presence)
            individual_metrics_data['consistency'].append(consistency)

            all_metrics['bertscore_f1'].append(bert_score)
            all_metrics['answer_presence'].append(answer_presence)
            all_metrics['consistency'].append(consistency)

        model_metrics = {
            'model': model_type.upper(),
            'bertscore_f1': sum(all_metrics['bertscore_f1']) / len(all_metrics['bertscore_f1']),
            'answer_presence': sum(all_metrics['answer_presence']) / len(all_metrics['answer_presence']),
            'consistency': sum(all_metrics['consistency']) / len(all_metrics['consistency'])
        }

        average_metrics_data['model'].append(model_metrics['model'])
        average_metrics_data['bertscore_f1'].append(model_metrics['bertscore_f1'])
        average_metrics_data['answer_presence'].append(model_metrics['answer_presence'])
        average_metrics_data['consistency'].append(model_metrics['consistency'])

    individual_metrics_df = pd.DataFrame(individual_metrics_data)
    average_metrics_df = pd.DataFrame(average_metrics_data)

    return individual_metrics_df, average_metrics_df
```

```python
# Disable transformers warnings
logging.set_verbosity_error()

with open('reference_answer.json', 'r') as file:
    reference_data = json.load(file)

reference_ans = {item['question_id']: item['reference_answer'] for item in reference_data}

individual_metrics_df, average_metrics_df = evaluate_models(responses_df, model_types, reference_ans)
```

```python
def plot_metrics(individual_df):
    mean_metrics_df = individual_df.groupby('model')[['bertscore_f1', 'answer_presence', 'consistency']].mean()

    ax = mean_metrics_df.plot(kind='bar', figsize=(10, 6))

    for p in ax.patches:
        ax.annotate(f'{p.get_height():.4f}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=12, color='black',
                    xytext=(0, 5), textcoords='offset points')

    plt.title("Average Performance of Models Across Metrics")
    plt.ylabel("Score")
    plt.xlabel("Model")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def summary_statistics(individual_df):
    summary_df = individual_df.groupby('model')[['bertscore_f1', 'answer_presence', 'consistency']].agg(['mean', 'std'])
    print("Summary Statistics:")
    print(summary_df)
```
